#Find and remove duplicates

```
#finding duplicates

from pyspark.sql.functions import col

duplicate_rows = df.groupBy(df.columns).count().filter(col("count")>1)
duplicate_rows.show()
```

```
#removing duplicates

duplicates_removed_df = df.dropDuplicates()
duplicates_removed_df.show()
```

```
#checking if duplicates are removed successfully

check_df = duplicates_removed_df.groupBy(duplicates_removed_df.columns).count().filter(col("count")>1)
check_df.show()
```
